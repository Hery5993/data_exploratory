```python
# Imports
import numpy as np
```

```python
arrA = np.array([1, np.nan, 3])
```

```python
# Problem when encouter NaN values
np.std(arrA)
```

```python
# To ignore NaN values we have to use adapted functions
# Such: https://numpy.org/doc/stable/reference/generated/numpy.nanstd.html
np.nanstd(arrA)
```
