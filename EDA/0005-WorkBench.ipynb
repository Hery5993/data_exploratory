```python
import numpy as np 
np.random.seed(93)
```

```python
""" Multiplication of matrix 
    use @ operator or np.matmul
    not * (*: will do multiplication element * element !!)

    cf: https://numpy.org/doc/2.2/reference/generated/numpy.matmul.html
    conditions must be filled: (n,k),(k,m)->(n,m)
"""
matA = np.random.randint(0, 100, size=6)
matA = matA.reshape((3, 2))
matA
```

```python
matB = np.random.randint(0, 50, size=6)
matB = matB.reshape((2, 3))
matB
```

```python
matC = matA @ matB
matC
```

```python
matD = matB @ matA
matD
```

```python
""" Transpose of matrix 
    - getting as attribute of ndarray object
    - by calling np.transpose (cf: https://numpy.org/doc/2.1/reference/generated/numpy.matrix.transpose.html)
"""
matA.T # getting as attribute of ndarray
```

```python
matE = np.transpose(matA) # by calling nb.transpose of NumPy
matE
```

```python
arrA = np.array([1, np.nan, 3])
```

```python
# Invalid result when encouter NaN values
print(np.std(arrA))
print(np.mean(arrA))
print(np.var(arrA))
```

```python
# To ignore NaN values we have to use adapted functions
# Such: https://numpy.org/doc/stable/reference/generated/numpy.nanstd.html
print(np.nanstd(arrA))
print(np.nanmean(arrA))
print(np.nanvar(arrA))
```

```python
sample_data = np.array([
    [12.4, 15.2, np.nan],
    [18.1, np.nan, 20.5]
])

nan_mask = np.isnan(sample_data)
```

```python
sample_data[nan_mask] = 10
sample_data
```
